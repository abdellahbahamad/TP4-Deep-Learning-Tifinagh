```python
import os
import numpy as np
from matplotlib import pyplot as plt
import cv2
import tensorflow.keras as keras  # Correction de l'importation
from keras.models import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

"""## Chargement et adaptation des données"""

def upload_data(path_name, number_of_images_per_class):
    X_Data = []
    Y_Data = []
    if not os.path.exists(path_name):
        print(f"Erreur : Le chemin n'existe pas : {path_name}")
        return np.array([]), np.array([])
    print(f"Analyse du chemin : {path_name}")
    # Obtenir dynamiquement tous les sous-dossiers comme catégories
    categories = [d for d in os.listdir(path_name) if os.path.isdir(os.path.join(path_name, d))]
    if not categories:
        print(f"Aucun sous-dossier trouvé dans {path_name}")
        return np.array([]), np.array([])
    print(f"Catégories détectées : {categories}")
    for idx, category in enumerate(categories):
        category_path = os.path.join(path_name, category)
        if os.path.isdir(category_path):
            images = [f for f in os.listdir(category_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
            print(f"Trouvé {len(images)} images dans {category}")
            if images:
                images_to_use = images[:min(number_of_images_per_class, len(images))]  # Limiter au nombre d'images disponibles
                for img_file in images_to_use:
                    img_path = os.path.join(category_path, img_file)
                    img = cv2.imread(img_path, 0)  # Charger en niveaux de gris
                    if img is not None:
                        # Redimensionner l'image à 32x32 immédiatement
                        img_resized = cv2.resize(img, (32, 32))
                        X_Data.append(img_resized)
                        Y_Data.append(idx)
                    else:
                        print(f"Échec du chargement de l'image : {img_path}")
                print(f"> Chargé {len(images_to_use)} images de la classe {category}.", end='\r')
            else:
                print(f"Aucune image trouvée dans {category}")
        else:
            print(f"Dossier non trouvé pour la catégorie : {category}")
    return np.array(X_Data) if X_Data else np.array([]), np.array(Y_Data) if Y_Data else np.array([])

# Définir les limites d'échantillonnage
n_train = 200  # Réduit à un nombre sûr (ajuster selon la taille du dataset, par ex. ~854 par classe)
n_test = 50    # Réduit à un nombre sûr (par ex. ~214 par classe)

# Mettre à jour les chemins pour correspondre à l'environnement Windows
train_path = r"C:\Users\Dell\Desktop\Tifinagh\train_data"
test_path = r"C:\Users\Dell\Desktop\Tifinagh\test_data"

# Charger les données d'entraînement et de test
print("Chargement des données d'entraînement...")
x_train, y_train = upload_data(train_path, n_train)
print("\nChargement des données de test...")
x_test, y_test = upload_data(test_path, n_test)

print("\nLa forme de x_train est :", x_train.shape)
print("La forme de x_test est :", x_test.shape)
print("La forme de y_train est :", y_train.shape)
print("La forme de y_test est :", y_test.shape)

def plot_data(num=3):
    if len(x_test) == 0:
        print("Erreur : x_test est vide, impossible d'afficher les données.")
        return
    fig, axes = plt.subplots(1, num, figsize=(12, 8))
    for i in range(num):
        index = np.random.randint(len(x_test))
        axes[i].imshow(x_test[index], cmap='gray')  # Utiliser directement, déjà 32x32
        axes[i].set_title('Étiquette de l\'image : %d' % y_test[index])
        axes[i].axis('off')
    plt.show()

if x_test.size > 0:
    plot_data(num=5)
    plot_data(num=5)

num_classes = len([d for d in os.listdir(test_path) if os.path.isdir(os.path.join(test_path, d))]) if os.path.exists(test_path) else 0  # Nombre de classes dynamique
size = 32  # Ajusté à 32x32 selon les exigences du TP

# Prétraiter les données
if x_train.size > 0 and x_test.size > 0:
    x_train = x_train.astype('float32')
    x_test = x_test.astype('float32')
    x_train = np.reshape(x_train, (x_train.shape[0], size * size))  # Déjà 32x32, aplatir directement
    x_test = np.reshape(x_test, (x_test.shape[0], size * size))    # Déjà 32x32, aplatir directement
    x_train /= 255
    x_test /= 255
    print('Forme de x_train :', x_train.shape)
    print(x_train.shape[0], 'échantillons d\'entraînement')
    print(x_test.shape[0], 'échantillons de test')

    # Convertir les vecteurs de classes en matrices de classes binaires
    y_train = to_categorical(y_train, num_classes)
    y_test = to_categorical(y_test, num_classes)

    """## Définir notre modèle de réseau de neurones (Architecture)"""

    model = Sequential()
    model.add(Dense(512, input_shape=(size * size,), activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(num_classes, activation='softmax'))

    model.compile(loss=keras.losses.categorical_crossentropy,
                  metrics=['accuracy'])

    model.summary()

    """## Prédiction du modèle sur les données de test avant entraînement"""

    def plot_predictions(model, num=3):
        if len(x_test) == 0:
            print("Erreur : x_test est vide, impossible d'afficher les prédictions.")
            return
        fig, axes = plt.subplots(1, num, figsize=(12, 8))
        for i in range(num):
            index = np.random.randint(len(x_test))
            pred = np.argmax(model.predict(np.reshape(x_test[index], (1, size * size))))
            axes[i].imshow(np.reshape(x_test[index], (size, size)), cmap='gray')
            axes[i].set_title('Étiquette prédite : ' + str(pred) + '\nÉtiquette réelle : ' + str(np.argmax(y_test[index])))
            axes[i].axis('off')
        plt.show()

    plot_predictions(model, num=5)

    """## Entraînement"""

    history = model.fit(x_train, y_train, batch_size=128, epochs=20, validation_data=(x_test, y_test))

    """## Prédiction du modèle sur les données de test après entraînement"""

    plot_predictions(model, num=5)
    score = model.evaluate(x_test, y_test, verbose=0)
    print('Perte sur les tests :', score[0])
    print('Précision sur les tests :', score[1])

    """## Historique du modèle pendant l'entraînement"""

    # Supprimé le style xkcd pour éviter les avertissements de police
    plt.plot(history.history['accuracy'], color='c')
    plt.plot(history.history['val_accuracy'], color='red')
    plt.title('Précision du modèle Tifinagh')
    plt.legend(['préc', 'val_préc'])
    plt.savefig(os.path.join(r"C:\Users\Dell\Desktop\Tifinagh", 'acc_Tifinagh.png'))
    plt.show()

    plt.plot(history.history['loss'], color='c')
    plt.plot(history.history['val_loss'], color='red')
    plt.title('Perte du modèle Tifinagh')
    plt.legend(['perte', 'val_perte'])
    plt.savefig(os.path.join(r"C:\Users\Dell\Desktop\Tifinagh", 'loss_Tifinagh.png'))
    plt.show()
else:
    print("Erreur : Données insuffisantes pour procéder à l'entraînement du modèle.")
```
